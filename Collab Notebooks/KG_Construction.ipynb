#### Naming conventions for neo4J
**Node labels**

Camel-case, beginning with an upper-case character

:VehicleOwner rather than :vehicle_owner etc.

**Relationship types**

Upper-case, using underscore to separate words

:OWNS_VEHICLE rather than :ownsVehicle etc.



## 1. Creation of Node Instance

```python
# Define the function to create a node with properties
def load_data(entity, properties):
    count = 0
    query = f"CREATE (n:{entity} $properties)"
    with driver.session() as session:
      for indvProp in properties:
        session.run(query, properties=indvProp)
        count+=1
        print(count)
```

```python
def create_rs(query, batch_size=100):
    with driver.session() as session:
        result = session.run(query)
        count = 0

        for record in result:
            count += 1
            # Print a message every 100 queries executed
            if count % batch_size == 0:
                print(f"Processed {count} queries")

        print(f"Total {count} queries processed")
```
